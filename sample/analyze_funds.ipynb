## 基準価額の推移

選択したいくつかのファンドのある期間での基準価額の推移を表示します。

```python
import datetime

funds = ['9C31116A', '29316153', '89311164']
start_period = datetime.date(2017, 1, 1)
end_period = datetime.date(2017, 6, 30)

em.show_price_chart(funds, start_period, end_period)
```

## 収益率の推移

選択したいくつかのファンドのある期間での収益率の推移を表示します。

```python
investment_period_days = 5

em.show_rate_of_return_chart(funds, start_period, end_period, investment_period_days)
```

## 平均-標準偏差ダイアグラム

複数選択されたファンドを収益率の平均-標準偏差ダイアグラム上で示します。
ランダムにファンドに重み付けしたときのポートフォリオ全体の平均-標準偏差を多数プロットすることにより、選択されたファンドに任意の重みで資産配分した場合のポートフォリオの実現可能領域を表します。


```python
num_random_feasible_set = 10000

em.show_mean_std_diagram(funds, start_period, end_period, investment_period_days, num_random_feasible_set)
```
